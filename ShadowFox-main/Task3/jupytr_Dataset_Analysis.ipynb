```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#we have to load and organize data so that it get organized and understanded by jupyter notebook
top_books_data = pd.read_csv('top_15_best_selling_books.csv')

print(top_books_data.head()) #use to display some of the starting rows
print("-------------------------------------------------------------------------------------------------------------------------------------")
print(top_books_data.info()) #use to find memory space/check null entries in csv file dataset
print("-------------------------------------------------------------------------------------------------------------------------------------")
print(top_books_data.describe()) #use to summarize the numerical columns and statistics 
print("-------------------------------------------------------------------------------------------------------------------------------------")

#Data cleaning-(drop missing values if any)
print(top_books_data.dropna())
```

```python
#To define a research question
print("1) How do sales for original prive V/S discounted price made a difference in between several ratings of top selling books.")
print("2) How does type of book whether be fiction, novel, core etc. made correletion between ratings of top selling books.")
```

```python
#Vizualize dataset for analysis --------- Original price vs discounted price effect on ratings
sns.set(style="whitegrid")
# Create a bar plot for Original and Discounted Prices
plt.figure(figsize=(12, 6))
bar_width = 0.35
index = range(len(top_books_data))
bars1 = plt.bar(index, top_books_data['Original Price'], width=bar_width, label='Original Price', color='lightblue')
bars2 = plt.bar([i + bar_width for i in index], top_books_data['Discounted Price'], width=bar_width, label='Discounted Price', color='salmon')
plt.scatter(index, top_books_data['Rating']*2, color='black', label='Rating (scaled x2)', s=100, zorder=5)
# Add labels and title
plt.xlabel('Books')
plt.ylabel('Price')
plt.title('Original Price and Discounted Price of Top 15 Best-Selling Books with Ratings')
plt.xticks([i + bar_width / 2 for i in index], top_books_data['Item'], rotation=45, ha='right')
plt.legend()
# Show the plot
plt.tight_layout()
plt.show()
```

```python
#Vizualize dataset for analysis ----------------- type of book effect on rating, people like which one more.
total_ratings_by_type = top_books_data.groupby('Type')['Total Ratings'].sum()
average_ratings_by_type = top_books_data.groupby('Type')['Rating'].mean()
average_ratings_by_type
# Tabular representation to show ratings.
popularity_df = pd.DataFrame({
    'Average Rating': average_ratings_by_type,
    'Total Ratings': total_ratings_by_type
})
popularity_df
```
